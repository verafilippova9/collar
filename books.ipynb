```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/books.csv')

month_map = {
    'Январь': 1, 'Февраль': 2, 'Март': 3, 'Апрель': 4, 'Май': 5, 'Июнь': 6,
    'Июль': 7, 'Август': 8, 'Сентябрь': 9, 'Октябрь': 10, 'Ноябрь': 11, 'Декабрь': 12
}

df['Месяц'] = df['Месяц'].map(month_map)


df_2023 = df[df['Год'] == 2023]
df_2024 = df[df['Год'] == 2024]

def plot_books_2023():
    plt.plot(df_2023['Месяц'], df_2023['Книги'], marker='o', color='red', label='Прочитано книг')
    plt.title("Прочитанные книги 2023")
    plt.xlabel("Месяцы")
    plt.ylabel("Количество книг")
    plt.grid(True)
    plt.xticks(df_2023['Месяц'], ['Январь', 'Февраль', 'Март', 'Апрель', 'Май', 'Июнь', 'Июль'])
    plt.legend()

def plot_pages_2024():
    plt.bar(df_2024['Месяц'], df_2024['Страницы'], color='green', label='Прочитано страниц')
    plt.title("Прочитанные страницы 2024")
    plt.xlabel("Месяцы")
    plt.ylabel("Количество страниц")
    plt.grid(True)
    plt.xticks(df_2024['Месяц'], ['Январь', 'Февраль', 'Март', 'Апрель', 'Май', 'Июнь', 'Июль', 'Август'])
    plt.legend()

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plot_books_2023()

plt.subplot(1, 2, 2)
plot_pages_2024()

plt.tight_layout()
plt.show()
```
